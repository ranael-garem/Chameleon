```python
import matplotlib.pyplot as plt
import numpy as np

bruno = 'Bruno'
jen = 'Jen'
tomislav = 'Tomislav'
joseph = 'Joseph'
kimberly = 'Kimberly'
alicia = 'Alicia'
rana = 'rana'

questions = [
    {
        'title': 'Would you rather eat pizza every single day or never eat pizza again?',
        'answers':
            {
                'Never': [bruno, tomislav, jen, kimberly, rana],
                'Every day': []
            }
    },
    {
        'title': 'Would you rather be a kid your whole life or an adult your whole life?',
        'answers':
            {
                'Kid': [bruno, tomislav, jen, rana],
                'Adult': [kimberly]
            }
    },
    {
        'title': 'Would you rather be able to read minds or be able to see the future?',
        'answers':
            {
                'See Future': [bruno, tomislav],
                'Read Minds': [kimberly, jen, rana]
            }
    },
    {
        'title': 'Would you rather be able to talk with animals or speak all foreign languages?',
        'answers':
            {
                'Animals': [bruno],
                'All Foreign Languages': [tomislav, kimberly, jen, rana]
            }
    },
    {
        'title': 'Would you rather have 10 dogs or no dogs at all?',
        'answers':
            {
                '10 Dogs': [bruno],
                'No Dogs': [tomislav, kimberly, jen, rana]
            }
    },
    {
        'title': 'Would you rather go deep sea diving or bungee jumping?',
        'answers':
            {
                'Deep Sea Diving': [bruno, tomislav, jen, kimberly],
                'Bungee Jumping': [rana]
            }
    },
    {
        'title': 'Would you rather live in the mountains or at the beach?',
        'answers':
            {
                'Mountains': [bruno, tomislav, rana],
                'Beach': [kimberly, jen]
            }
    },
    {
        'title': 'Would you rather mentally or physically never age?',
        'answers':
            {
                'Mentally': [kimberly, jen],
                'Physically': [bruno, tomislav, rana],
            }
    },
    {
        'title': 'Would you rather lose all your technology devices or lose every picture thatâ€™s ever been taken of you?',
        'answers':
            {
                'Pictures': [bruno, jen, rana],
                'Tech Devices': [kimberly, tomislav]
            }
    }
]
```

```python
for question in questions:
    title = question['title']
    answers = question['answers']
    left = list(answers.keys())
    height = []
    for item in left:
        current_answer = answers[item]
        height.append(len(current_answer))



    # # plotting a bar chart
    plt.bar(left, height,
            width=0.8, color=['red', 'green'])
    plt.yticks(np.arange(0, 7, 1))


    plt.title(title)
    plt.show()
```
